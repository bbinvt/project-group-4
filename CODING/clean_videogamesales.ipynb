```python
import pandas as pd
vgdata = pd.read_csv('https://sbcharitybucket.s3.us-west-2.amazonaws.com/Video_Games.csv')
vgdata.head()
```

```python
# Check data types
vgdata.dtypes
```

```python
# Drop unnecessary/problematic columns
vgdata = vgdata.drop(['Developer','User_Count','Critic_Count'], axis = 1)
```

```python
vgdata.head()
```

```python
# Check for Null values
vgdata.isnull().sum()
```

```python
# Drop null values
vgdata = vgdata.dropna()
```

```python
vgdata.isnull().sum()
```

```python
vgdata.describe()
```

```python
# Normalise Critic Score
vgdata['Critic_Score'] = vgdata['Critic_Score']/10
vgdata.head()
```

```python
# Normalise sales figures
vgdata['NA_Sales'] = vgdata['NA_Sales']*0.8
vgdata['EU_Sales'] = vgdata['EU_Sales']*0.8
vgdata['JP_Sales'] = vgdata['JP_Sales']*0.8
vgdata['Other_Sales'] = vgdata['Other_Sales']*0.8
vgdata['Global_Sales'] = vgdata['Global_Sales']*0.8
```

```python
vgdata.head()
```

```python
vgdata = vgdata.loc[vgdata['User_Score'] != 'tbd']
print(vgdata.shape)
```

```python
vgdata['User_Score'] = vgdata['User_Score'].astype(float)
vgdata.dtypes
```

```python
vgdata.to_csv('clean_videogamesales.csv', index = False, encoding = 'UTF8')
```
