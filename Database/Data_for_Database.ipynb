```python
# Import dependencies 
import pandas as pd 

# Read in csv file 
data = pd.read_csv('clean_videogamesales.csv')
```

```python
# Look at the data 
data.head()
```

```python
# Create DataFrame containing data contained in the games table 
games_df = pd.DataFrame(data, columns=["Name","Platform","Year_of_Release",
                                       "Genre","Publisher",
                                       "Developer","Rating"])

games_df.head()
```

```python
sales_df = pd.DataFrame(data, columns=["Name","NA_Sales","EU_Sales",
                                       "JP_Sales","Other_Sales",
                                       "Global_Sales"])

sales_df.head()
```

```python
ratings_df = pd.DataFrame(data, columns=["Name","Critic_Score",
                                         "Critic_Count",
                                         "User_Score", "User_Count"])

ratings_df.head()
```

```python
games_df.to_csv('Games.csv', index=False, sep='\t')
sales_df.to_csv('Sales.csv', index=False, sep='\t')
ratings_df.to_csv('Ratings.csv', index=False, sep='\t')
```
