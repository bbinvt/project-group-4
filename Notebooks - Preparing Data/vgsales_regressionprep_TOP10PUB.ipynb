```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

vgdata = pd.read_csv('../DATA/clean_videogamesales.csv')
vgdata.head()
```

```python
vgdata2 = vgdata.copy()
```

```python
vgdata2 = vgdata2.drop(['Name'], axis = 1)
```

```python
vgdata2['Years_on_Market'] = 2016 - vgdata2['Year_of_Release']
```

```python
vgdata2['Global_Sales'] = vgdata2['Global_Sales'] * (0.2*vgdata2['Years_on_Market'])
vgdata2['NA_Sales'] = vgdata2['NA_Sales'] * (0.2*vgdata2['Years_on_Market'])
vgdata2['EU_Sales'] = vgdata2['EU_Sales'] * (0.2*vgdata2['Years_on_Market'])
vgdata2['JP_Sales'] = vgdata2['JP_Sales'] * (0.2*vgdata2['Years_on_Market'])
vgdata2['Other_Sales'] = vgdata2['Other_Sales'] * (0.2*vgdata2['Years_on_Market'])
```

```python
vgdata2.head()
```

```python
vgdata2 = vgdata2.drop(['Year_of_Release'], axis = 1)
vgdata2.head()
```

```python
vgdata2.columns
```

```python
rat_counts = vgdata2['Rating'].value_counts()
rat_counts
```

```python
# Determine which values to replace if counts are less than ...?
replace_rating = list(rat_counts[rat_counts < 1000].index)

# Replace in dataframe
for rat in replace_rating:
    vgdata2.Rating = vgdata2.Rating.replace(rat,"Other")
    
# Check to make sure binning was successful
vgdata2.Rating.value_counts()
```

```python
vgdata2.head()
```

```python
vgdata2 = vgdata2[['Platform','Publisher','NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales','Years_on_Market','Critic_Score', 'User_Score','Genre', 'Rating']]
```

```python
vgdata2.head()
```

```python
plat_counts = vgdata2['Platform'].value_counts()
plat_counts
```

```python
# Determine which values to replace if counts are less than ...?
replace_platform = list(plat_counts[plat_counts < 150].index)

# Replace in dataframe
for plat in replace_platform:
    vgdata2.Platform = vgdata2.Platform.replace(plat,"Other")
    
# Check to make sure binning was successful
vgdata2.Platform.value_counts()
```

```python
pub_counts = vgdata2['Publisher'].value_counts()
pub_counts
```

```python
# Determine which values to replace if counts are less than ...?
replace_publisher = list(pub_counts[pub_counts < 200].index)

# Replace in dataframe
for pub in replace_publisher:
    vgdata2.Publisher = vgdata2.Publisher.replace(pub,"Other")
    
# Check to make sure binning was successful
vgdata2.Publisher.value_counts()
```

```python
vgdata2 = vgdata2.loc[vgdata2['Publisher'] != 'Other']
print(vgdata2.shape)
```

```python
vgdata2.Publisher.value_counts()
```

```python
vgdata2.info()
```

```python
# Generate our categorical variable lists
vgsales2_cat = vgdata2.dtypes[vgdata2.dtypes == "object"].index.tolist()
print(vgsales2_cat)
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df2 = pd.DataFrame(enc.fit_transform(vgdata2[vgsales2_cat]))

# Add the encoded variable names to the dataframe
encode_df2.columns = enc.get_feature_names(vgsales2_cat)
encode_df2.head()
```

```python
encode_df2.describe()
```

```python
vgdataCL = vgdata2.merge(encode_df2,left_index=True, right_index=True)
vgdataCL = vgdataCL.drop(vgsales2_cat,1)
vgdataCL.head()
```

```python
vgdataCL.describe()
```

```python
vgdataCL.to_csv('../DATA/regreadyvgsales_TOP10PUB.csv', index = False, encoding = "UTF8")
```
